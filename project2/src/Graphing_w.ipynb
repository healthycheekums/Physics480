```python
import numpy as np                  ## Imports Python's numerical library
import matplotlib.pyplot as plt     ## Imports matplotlib for plotting
%matplotlib inline

#varied velocity
# w = [2.8368,0.4451,1.240,2.1891,3.8802,17.645]
#E = [0,.01,.05,.25,.5,1,5]
E =[0.1106,0.1231,0.53431,0.54408,0.65019,0.96438,4.38549]
i = 0
r= np.arange(0,15,0.1)
plt.hold(True)
while i < len(E):
    wave = r*np.exp((E[i]/7)*-(r**2))
    plt.plot(r,wave,label="E="+str(E[i]))
    i +=1
plt.legend(loc='best')
plt.xlabel ('radius')
plt.ylabel ('Magnitude')
plt.title ('Varying Lamda')
plt.show()
```
